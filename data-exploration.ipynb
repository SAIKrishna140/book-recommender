```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
ax=plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)
plt.ylabel("Columns")
plt.xlabel("Missing Values")
plt.show()
```

```python
import numpy as np
books["missing_description"]=np.where(books["description"].isna(), 1, 0)
books["age_of_book"]=2025-books["published_year"]
```

```python
books.head()
```

```python
columns_of_interest=["average_rating","missing_description","age_of_book","num_pages"]

correlation_matrix=books[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap=sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm",fmt=".2f",cbar_kws={"label": "SpearMan Correlation"})
heatmap.set_title("Correlation Matrix")
plt.show()
```

```python
books_missing=books[(~books["num_pages"].isna())&(~books["description"].isna())&(~books["average_rating"].isna())&(~books["published_year"].isna())]
```

```python
books_missing
```

```python
books_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
books_missing.loc[:,"words_in_description"]=books_missing["description"].str.split().str.len()
```

```python
books_missing
```

```python
books_missing.loc[books_missing["words_in_description"].between(25,34), "description"]
```

```python
books_missing_25_words=books_missing[books_missing["words_in_description"]>=25]
```

```python
books_missing_25_words
```

```python
books_missing_25_words.loc[:,"title_and_subtitle"]=(np.where(books_missing_25_words["subtitle"].isna(),books_missing_25_words["title"],books_missing_25_words[["title","subtitle"]].astype(str).agg(":".join,axis=1)))
```

```python
books_missing_25_words
```

```python
books_missing_25_words.loc[:,"tagged_description"]=books_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join,axis=1)
```

```python
books_missing_25_words
```

```python
(books_missing_25_words.drop(["subtitle","missing_description","age_of_book","words_in_description"],axis=1)
 .to_csv("books_cleaned_data.csv",index=False))
```
